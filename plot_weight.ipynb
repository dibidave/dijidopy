```python
import notion
from plotly import express as plotly_express
import pandas
from datetime import timedelta
```

```python
df = notion.get_weight_log()
```

```python
df.sort_values("date", inplace=True)
```

```python
figure = plotly_express.scatter(df, x="date", y="weight")
```

```python
# Add a column for a 7-day moving average
df["7-day moving average"] = df["weight"].rolling(window=7, center=True).mean()

# Add a trendline for the 7-day moving average  
figure.add_scatter(x=df["date"], y=df["7-day moving average"], mode="lines", name="7-day moving average")
```

```python
# Find the smallest moving average window size that creates a monotonically decreasing trendline
for window_size in range(1, len(df)):
    df[f"{window_size}-day moving average"] = df["weight"].rolling(window=window_size, center=True).mean()
    if df[~pandas.isna(df[f"{window_size}-day moving average"])][f"{window_size}-day moving average"].is_monotonic_decreasing:
        figure.add_scatter(x=df["date"], y=df[f"{window_size}-day moving average"], mode="lines", name=f"{window_size}-day moving average")
        break
```
