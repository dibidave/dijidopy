```python
import dijido
from datetime import timedelta
from datetime import datetime
import pytz
import pandas
import numpy

from plotly import graph_objects
from plotly import offline as plotly

import plotly.io as pio
pio.renderers.default = "notebook"

import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)

import scipy
from scipy import signal
```

```python
transactions = dawlla.get_transactions()
accounts = dawlla.get_accounts()
categories = dawlla.get_categories()
parties = dawlla.get_legal_entities()
```

```python
amount_by_party = {
    party["_id"]: 0 for party in parties
}

for transaction in transactions:
    party_id = transaction["party_id"]
    amount = transaction["amount"]
    amount_by_party[party_id] += amount


# Make a table of the parties by amount spent, looking up the name of the party
# From the parties list
party_names = {
    party["_id"]: party["name"] for party in parties
}

party_amounts = [
    {
        "party_id": party_id,
        "party_name": party_names[party_id],
        "amount": amount
    }
    for party_id, amount in amount_by_party.items()
]

party_amounts = sorted(
    party_amounts,
    key=lambda x: x["amount"],
    reverse=False
)

party_amounts = pandas.DataFrame(party_amounts)
```
