```python
import os
from datetime import datetime
from datetime import timezone
from datetime import timedelta
```

```python
david_sleep_intervals = []

with open("com.samsung.health.sleep_stage.20230817234201.csv") as sleep_file:
    
    sleep_file.readline()
    sleep_file.readline()
    
    while True:
        
        line = sleep_file.readline()
        
        if not line:
            break
        
        line_elements = line.strip().split(",")
        start_time_string = line_elements[0]
        end_time_string = line_elements[9]
        
        start_time = datetime.strptime(start_time_string, "%Y-%m-%d %H:%M:%S.%f").replace(tzinfo=timezone.utc)
        end_time = datetime.strptime(end_time_string, "%Y-%m-%d %H:%M:%S.%f").replace(tzinfo=timezone.utc)
        
        stage = line_elements[5]
        
        if stage == "40001":
            stage = "Awake"
        elif stage == "40002":
            stage = "Light"
        elif stage == "40003":
            stage = "Deep"
        elif stage == "40004":
            stage = "REM"
        else:
            raise ValueError(f"Samsung you lied to me, {stage} is not a stage!")
        
        david_sleep_intervals.append((start_time, end_time, stage))
        
tatyana_sleep_times = []

with open("tatyana_sleep_log.csv") as sleep_file:
    
    sleep_file.readline()
    
    while True:
        
        line = sleep_file.readline()
        
        if not line:
            break
        
        line_elements = line.strip().split("\t")
        date_string = line_elements[0]
        minutes = line_elements[1]
        stage = line_elements[2]
        
        date = datetime.strptime(date_string + " 23:59", "%m/%d/%Y %H:%M").replace(tzinfo=datetime.now().astimezone().tzinfo)
        
        tatyana_sleep_times.append((date, minutes, stage))
```

```python
sleep_data = []

with open("light_log.csv") as light_file:
    
    light_file.readline()
    
    while True:
        
        line = light_file.readline()
        
        if not line:
            break
        
        line_elements = line.strip().split(",")
        date_string = line_elements[0]
        is_light_on = line_elements[1]
        
        if is_light_on == "True":
            is_light_on = True
        else:
            is_light_on = False
        
        date = datetime.strptime(date_string + " 16:00", "%Y-%m-%d %H:%M").replace(tzinfo=datetime.now().astimezone().tzinfo)
        end_date = date + timedelta(days=1)
        
        # Find all sleep entries in this date range
        
        sleep_stage_sums = {}
        
        for interval in david_sleep_intervals:
            
            if interval[0] < date or interval[0] > end_date:
                continue
            
            sleep_stage = interval[2]
            
            if sleep_stage not in sleep_stage_sums:
                sleep_stage_sums[sleep_stage] = 0
            
            sleep_stage_sums[sleep_stage] += (interval[1] - interval[0]).total_seconds() / 60
        
        if sum(sleep_stage_sums.values()) > 0:
            for sleep_stage, minutes in sleep_stage_sums.items():

                sleep_data.append({
                    "date": date_string,
                    "sleep_stage": sleep_stage,
                    "minutes": minutes,
                    "ratio": minutes / sum(sleep_stage_sums.values()),
                    "is_light_on": is_light_on,
                    "weekday": date.strftime('%A'),
                    "participant": "David"
                })

            sleep_data.append({
                "date": date_string,
                "sleep_stage": "Total",
                "minutes": sum(sleep_stage_sums.values()),
                "ratio": 1,
                "is_light_on": is_light_on,
                "weekday": date.strftime('%A'),
                "participant": "David"
            })
        
        sleep_stage_sums = {}
        
        for stage_time in tatyana_sleep_times:
            
            if stage_time[0] < date or stage_time[0] > end_date:
                continue
            
            sleep_stage = stage_time[2]
            
            if sleep_stage not in sleep_stage_sums:
                sleep_stage_sums[sleep_stage] = 0
            
            sleep_stage_sums[sleep_stage] += int(stage_time[1])
           
        if sum(sleep_stage_sums.values()) > 0: 
            for sleep_stage, minutes in sleep_stage_sums.items():

                sleep_data.append({
                    "date": date_string,
                    "sleep_stage": sleep_stage,
                    "minutes": minutes,
                    "ratio": minutes / sum(sleep_stage_sums.values()),
                    "is_light_on": is_light_on,
                    "weekday": date.strftime('%A'),
                    "participant": "Tatyana"
                })

            sleep_data.append({
                "date": date_string,
                "sleep_stage": "Total",
                "minutes": sum(sleep_stage_sums.values()),
                "ratio": 1,
                "is_light_on": is_light_on,
                "weekday": date.strftime('%A'),
                "participant": "Tatyana"
            })
```

```python
sleep_data = []

current_date = datetime.strptime("2021-09-24 16:00", "%Y-%m-%d %H:%M").replace(tzinfo=datetime.now().astimezone().tzinfo)

while current_date < datetime.now().replace(tzinfo=datetime.now().astimezone().tzinfo):
    
    end_date = current_date + timedelta(days=1)
    
    date_string = current_date.strftime("%Y-%m-%d")
    
    print(date_string)

    # Find all sleep entries in this date range

    sleep_stage_sums = {}

    for interval in david_sleep_intervals:

        if interval[0] < current_date or interval[0] > end_date:
            continue

        sleep_stage = interval[2]

        if sleep_stage not in sleep_stage_sums:
            sleep_stage_sums[sleep_stage] = 0

        sleep_stage_sums[sleep_stage] += (interval[1] - interval[0]).total_seconds() / 60

    if sum(sleep_stage_sums.values()) > 0:
        for sleep_stage, minutes in sleep_stage_sums.items():

            sleep_data.append({
                "date": date_string,
                "sleep_stage": sleep_stage,
                "minutes": minutes,
                "ratio": minutes / sum(sleep_stage_sums.values()),
                "is_light_on": is_light_on,
                "weekday": current_date.strftime('%A'),
                "month": current_date.month,
                "participant": "David"
            })

        sleep_data.append({
            "date": date_string,
            "sleep_stage": "Total",
            "minutes": sum(sleep_stage_sums.values()),
            "ratio": 1,
            "is_light_on": is_light_on,
            "weekday": current_date.strftime('%A'),
            "month": current_date.month,
            "participant": "David"
        })
        
    current_date = end_date
```
