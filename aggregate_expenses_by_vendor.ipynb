```python
import pandas
pandas.options.display.max_rows = 1000
pandas.options.display.min_rows = 100
```

```python
dfs = []

files_parameters = {
    "dawlla_expenses_cybernoodlers.xlsx": ("Sheet1", "party_name", "amount", 0.5),
    "dawlla_expenses_dibidave.xlsx": ("Sheet1", "party_name", "amount", 1),
    "David\'s Finances.xlsx": ("Expenses", "Vendor", "Amount", -1),
    "Finances.xlsx": ("Expenses", "From", "Amount", 0.5)
}

for file_name, parameters in files_parameters.items():

    sheet_name, party_name_col, amount_col, multiplier = parameters
    df = pandas.read_excel(file_name, sheet_name=sheet_name)
    df[amount_col] *= multiplier
    df = df[[party_name_col, amount_col]]
    df.columns = ["party_name", "amount"]
    dfs.append(df)

# Sum up all the amounts by party name
party_amounts = pandas.concat(dfs).groupby("party_name").sum().reset_index()
```

```python
# Sort the DataFrame by amount in descending order
party_amounts = party_amounts.sort_values(by="amount", ascending=True)
party_amounts
```
