```python
datas = []

for file in os.listdir("."):
    
    if not file.endswith(".json"):
        continue
    
    data = json.load(open(file))
    
    for entry in data:
        
        datas.append({
            "weight": entry["weight"],
            "date": entry["date"]
        })
```

```python
import pandas
df = pandas.DataFrame.from_records(datas)
df.to_excel("weight.xlsx")
```
